```python
import numpy as np
```

```python
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5], #np.nan is null value
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
#Removing missing Data
```

```python
#This works on rows
```

```python
df.dropna()
```

```python
df.dropna(thresh=3) #There should be at least three null values
```

```python
#Filling the Missing Data
```

```python
df.fillna(0)
```

```python
#fillna happens colunm wise
```

```python
values={'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
