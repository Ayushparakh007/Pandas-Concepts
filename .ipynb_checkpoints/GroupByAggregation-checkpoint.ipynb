```python
import numpy as np
import pandas as pd
```

```python

data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)

```

```python
df
```

```python
# Group by Category and calculate the sum of Sales
```

```python
df.groupby('Category')
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
# Group by Store and calculate the sum of Sales
df.groupby('Store')['Sales'].sum()
```

```python
# Group by multiple columns
# Group by Category and Store
df.groupby(['Category','Store'])['Sales'].sum()
```

```python
#Aggregation combing multiple value
```

```python
df['Sales'].mean() 
"""mean medain mode min max count std"""
```

```python
df['Sales'].agg(['mean','median','min','max'])
```
